Необходимо разработать класс `Pet`, который будет представлять питомца в контексте системы управления отелем для животных. В классе должны быть инкапсулированы следующие свойства:

1. **name**: имя питомца.
2. **breed**: порода питомца.
3. **age**: возраст питомца.

Для каждого из этих свойств требуется реализовать методы доступа (геттеры) и методы изменения (сеттеры). Для этого используйте декораторы `@property` и `@<property_name>.setter`.

#### Требования:
1. Создайте класс `Pet`, который принимает три параметра в конструкторе: `name`, `breed` и `age`.
2. Реализуйте геттеры для каждого из этих свойств с помощью декоратора `@property`.
3. Реализуйте сеттеры для каждого из этих свойств с помощью декоратора `@<property_name>.setter`.
4. Продемонстрируйте работу класса, создав объект `Pet` с определёнными значениями и изменив его свойства через сеттеры.



```python
class Pet:
    def __init__(self, name, breed, age):
        self._name = name
        self._breed = breed
        self._age = age

    @property
    def name(self):
        """Возвращает имя питомца."""
        return self._name

    @name.setter
    def name(self, new_name):
        """Устанавливает новое имя питомца."""
        self._name = new_name

    @property
    def breed(self):
        """Возвращает породу питомца."""
        return self._breed

    @breed.setter
    def breed(self, new_breed):
        """Устанавливает новую породу питомца."""
        self._breed = new_breed

    @property
    def age(self):
        """Возвращает возраст питомца."""
        return self._age

    @age.setter
    def age(self, new_age):
        """Устанавливает новый возраст питомца."""
        if isinstance(new_age, int) and new_age > 0:
            self._age = new_age
        else:
            print("Возраст должен быть положительным целым числом.")

# Пример использования:
pet = Pet("Лара", "Кошка", 3)
print(f"Имя: {pet.name}")
print(f"Порода: {pet.breed}")
print(f"Возраст: {pet.age}")

# Изменение свойств через сеттеры
pet.name = "Люси"
pet.age = 4
pet.breed = "Сиамская кошка"

print(f"\nОбновленное имя: {pet.name}")
print(f"Обновленная порода: {pet.breed}")
print(f"Обновленный возраст: {pet.age}")

# Попытка установить некорректный возраст
pet.age = -1  # Возраст должен быть положительным целым числом.
print(f"Возраст после некорректной установки: {pet.age}") # Возраст останется 4

pet.age = 0 # Возраст должен быть положительным целым числом.
print(f"Возраст после некорректной установки: {pet.age}") # Возраст останется 4

pet.age = 5
print(f"Возраст после корректной установки: {pet.age}") # Возраст станет 5
```

Создайте два класса-наследника Cat и Dog, которые будут добавлять уникальные атрибуты и методы для каждого типа животного.

#### Требования:

1. **Класс `Cat`**, который наследуется от `Pet`, должен добавлять:
   - **favorite_toy** (любимая игрушка).
   - Метод `play()`, который возвращает строку, описывающую игру питомца с любимой игрушкой.

2. **Класс `Dog`**, который также наследуется от `Pet`, должен добавлять:
   - **trick** (трик, который питомец умеет выполнять).
   - Метод `perform_trick()`, который возвращает строку с описанием выполнения трюка.

3. Использовать **сеттеры** для изменения значений, например, для установки любимой игрушки у кошки и выполнения трюка у собаки.



```python
class Pet:
    def __init__(self, name, breed, age):
        self._name = name
        self._breed = breed
        self._age = age

    @property
    def name(self):
        """Возвращает имя питомца."""
        return self._name

    @name.setter
    def name(self, new_name):
        """Устанавливает новое имя питомца."""
        self._name = new_name

    @property
    def breed(self):
        """Возвращает породу питомца."""
        return self._breed

    @breed.setter
    def breed(self, new_breed):
        """Устанавливает новую породу питомца."""
        self._breed = new_breed

    @property
    def age(self):
        """Возвращает возраст питомца."""
        return self._age

    @age.setter
    def age(self, new_age):
        """Устанавливает новый возраст питомца."""
        if isinstance(new_age, int) and new_age > 0:
            self._age = new_age
        else:
            print("Возраст должен быть положительным целым числом.")

class Cat(Pet):
    def __init__(self, name, breed, age, favorite_toy):
        super().__init__(name, breed, age) # Вызов конструктора родительского класса
        self._favorite_toy = favorite_toy

    @property
    def favorite_toy(self):
        """Возвращает любимую игрушку кошки."""
        return self._favorite_toy

    @favorite_toy.setter
    def favorite_toy(self, new_toy):
        """Устанавливает новую любимую игрушку для кошки."""
        self._favorite_toy = new_toy

    def play(self):
        """Возвращает строку, описывающую игру кошки с любимой игрушкой."""
        return f"{self.name} играет с {self.favorite_toy}"

class Dog(Pet):
    def __init__(self, name, breed, age, trick):
        super().__init__(name, breed, age) # Вызов конструктора родительского класса
        self._trick = trick

    @property
    def trick(self):
        """Возвращает трюк, который умеет выполнять собака."""
        return self._trick

    @trick.setter
    def trick(self, new_trick):
        """Устанавливает новый трюк для собаки."""
        self._trick = new_trick

    def perform_trick(self):
        """Возвращает строку с описанием выполнения трюка собакой."""
        return f"{self.name} выполняет трюк: {self.trick}"

# Пример использования:
cat = Cat("Лара", "Кошка", 3, "мячик")
dog = Dog("Рекс", "Немецкая овчарка", 5, "дает лапу")

print(f"Кошка: {cat.name}, Порода: {cat.breed}, Возраст: {cat.age}, Любимая игрушка: {cat.favorite_toy}")
print(cat.play())

cat.favorite_toy = "ниточка"
print(f"Новая любимая игрушка Лары: {cat.favorite_toy}")
print(cat.play())

print(f"\nСобака: {dog.name}, Порода: {dog.breed}, Возраст: {dog.age}, Трюк: {dog.trick}")
print(dog.perform_trick())

dog.trick = "крутит хвост"
print(f"Новый трюк Рекса: {dog.trick}")
print(dog.perform_trick())

cat.age = 4
dog.name = "Макс"
print(f"\nНовый возраст Лары: {cat.age}")
print(f"Новое имя собаки: {dog.name}")
```

Добавьте в классы Cat и Dog метод `make_sound()`, который будет выводить характерный звук для каждого животного.\
В методе `make_sound()` используйте полиморфизм с `@property`, чтобы каждый класс реализовал свой собственный звук.

#### Требования:

1. **Базовый класс `Pet`**:
   - Должен содержать атрибуты: `name` (имя питомца), `breed` (порода) и `age` (возраст).
   - Для каждого питомца должен быть предусмотрен метод `sound()`, который будет переопределяться в дочерних классах.
   - Метод `sound()` должен быть абстрактным (с использованием исключения `NotImplementedError`), что означает, что дочерние классы обязаны реализовать этот метод.

2. **Класс `Cat`, наследующий от `Pet`**:
   - Добавьте атрибут `favorite_toy` (любимая игрушка).
   - Переопределите метод `sound()`, чтобы выводить типичный для кошек звук, например, `"Meow!"`.
   - Добавьте метод `make_sound()`, который будет выводить строку: `"Имя питомца говорит: Meow!"` с использованием полиморфизма через `@property`.

3. **Класс `Dog`, наследующий от `Pet`**:
   - Добавьте атрибут `trick` (трик, который питомец умеет выполнять).
   - Переопределите метод `sound()`, чтобы выводить типичный для собак звук, например, `"Woof!"`.
   - Добавьте метод `make_sound()`, который будет выводить строку: `"Имя питомца говорит: Woof!"` с использованием полиморфизма через `@property`.

4. **Использование полиморфизма**:
   - В методе `make_sound()` примените полиморфизм с использованием `@property` для каждого класса, чтобы метод динамически определял, какой звук воспроизводит питомец в зависимости от его типа.




```python
class Pet:
    def __init__(self, name, breed, age):
        self._name = name
        self._breed = breed
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, new_name):
        self._name = new_name

    @property
    def breed(self):
        return self._breed

    @breed.setter
    def breed(self, new_breed):
        self._breed = new_breed

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, new_age):
        if isinstance(new_age, int) and new_age > 0:
            self._age = new_age
        else:
            print("Возраст должен быть положительным целым числом.")

    @property
    def sound(self):
        raise NotImplementedError("Свойство 'sound' должно быть реализовано в дочернем классе.")


class Cat(Pet):
    def __init__(self, name, breed, age, favorite_toy="неизвестная игрушка"):
        super().__init__(name, breed, age)
        self._favorite_toy = favorite_toy

    @property
    def favorite_toy(self):
        return self._favorite_toy

    @favorite_toy.setter
    def favorite_toy(self, new_toy):
        self._favorite_toy = new_toy

    def play(self):
        return f"{self.name} играет с {self.favorite_toy}"

    @property
    def sound(self):
        return f"{self.name} говорит: Meow!"

    def make_sound(self):
        return self.sound # self.sound является полиморфным свойством


class Dog(Pet):
    def __init__(self, name, breed, age, trick="неизвестный трюк"):
        super().__init__(name, breed, age)
        self._trick = trick

    @property
    def trick(self):
        return self._trick

    @trick.setter
    def trick(self, new_trick):
        self._trick = new_trick

    def perform_trick(self):
        return f"{self.name} выполняет трюк: {self.trick}"

    @property
    def sound(self):
        return f"{self.name} говорит: Woof!"

    def make_sound(self):
        return self.sound # self.sound является полиморфным свойством

# Пример использования:
cat = Cat("Lara", "Cat", 3)
dog = Dog("Rex", "Dog", 5)

print(cat.sound)
print(dog.sound)

# Демонстрация работы make_sound
print(cat.make_sound())
print(dog.make_sound())

# Можно также создать с указанием игрушки/трюка
cat_with_toy = Cat("Whiskers", "Siamese", 2, "лазерная указка")
dog_with_trick = Dog("Buddy", "Golden Retriever", 4, "приносит тапочки")

print(f"\n{cat_with_toy.name} (${cat_with_toy.breed}$) любимая игрушка: ${cat_with_toy.favorite_toy}")
print(cat_with_toy.sound)
print(cat_with_toy.play())

print(f"\n{dog_with_trick.name} (${dog_with_trick.breed}$) умеет трюк: ${dog_with_trick.trick}")
print(dog_with_trick.sound)
print(dog_with_trick.perform_trick())
```
